## 3. Bringing Music21 into pandas via the VIS-Framework

```python
# -*- coding: utf-8 -*-

'''Load some libraries first.'''
'''Bring in Music21.'''
import music21
'''Load some items from the vis-framework.'''
from vis.analyzers.indexers import noterest, interval
```

```python
'''Parse a score with Music21 into a music21 stream.'''
mto_score = music21.converter.parse('scores/symbolic/Iste-sanctus-pro-lege-Dei-sui_Victoria-Tomas-Luis-de.mid')
'''Place the Music21 stream into a pandas dataframe via the VIS-Framework.'''
vis_score = noterest.NoteRestIndexer(mto_score).run()
```

```python
'''Quickly viewing the newly created dataframe.'''
vis_score.fillna(' ^ ').tail(21)
```

Let's double check that with a graphic score.
![A visual representation of the last 21 events of Victoria's _Iste sanctus pro lege Dei sui_ (This holy place of God's law)](scores/graphic/excerpts/Victoria-Example.png "Last 21 events of Victoria's _Iste sanctus pro lege Dei sui_ (This holy place of god's law")

Perhaps we can look at the dataframe in a more "score-like" manner?

```python
'''Looking at the dataframe in a more score like format.'''
vis_score.fillna('<').tail(21).T
```

Hey, but what about those numbers that indicate the cadence? Can we have vertical intervals?

Sure can!

```python
'''Specify the Vertical INTerval indexer SETTINGS.'''
vint_settings = {
	'directed': False, 
	'simple or compound': 'simple', # Y not 'interval' : 'simple' || 'interval' : 'compound' ? 
	'quality': False, 
	'mp': False # 'mp' - such meaningless ... 'multi_process' - such self documenting ...
}

'''Now create a new dataframe consisting of vertical intervals.'''
vints = interval.IntervalIndexer(vis_score,vint_settings).run()
```

```python
'''Assigning only the supplementum to a variable.'''
supplementum = vints.tail(21)
'''Viewing the supplementum:'''
supplementum.fillna('')
```
